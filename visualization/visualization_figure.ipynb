```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker
import matplotlib.patches as patches
import numpy as np

new_data = {
    'Model': ['SGPN', 'SGFN', 'SGG point', 'SGFN+Attn', 'VL-SAT'],
    'Total_Params': [3.29, 2.90, 1.92, 8.20, 25.06],  # : 
    'Accuracy': [85.94, 88.53, 88.7, 88.7, 89.89],
    'FLOPs': [0.8637, 0.8413, 1.9700, 1.9700, 3.3722]  # : 
}
df_new = pd.DataFrame(new_data)

plt.style.use('seaborn-whitegrid')
plt.rcParams['font.family'] = 'serif'
fig, ax = plt.subplots(figsize=(5, 3))  
colors = ['gray', 'gray', 'gray', 'gray', 'gray']

def flops_scale(x):
    if x < 1.0:
        return 100 + (x - 0.8) / 0.2 * 200
    elif x < 2.0:
        return 300 + (x - 1.0) / 1.0 * 300
    else:
        return 800 + (x - 2.0) / 1.4 * 900

for i, (params, acc, model, flops, color) in enumerate(zip(df_new['Total_Params'], df_new['Accuracy'], df_new['Model'], df_new['FLOPs'], colors)):
    scaled_flops = flops_scale(flops)
    ax.scatter(params, acc, s=scaled_flops, c=color, edgecolors='black', linewidth=1, zorder=5, alpha=0.5)
    ax.annotate(f'{model}', (params, acc), xytext=(-1, 20), 
                textcoords='offset points', ha='center', va='bottom', fontsize=9)

ax.set_xlabel('Total Parameters (M)', fontsize=10)
ax.set_ylabel('Triplet Accuracy (%)', fontsize=10)
ax.set_xscale('log')
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'{x:.1f}M' if x < 1000 else f'{x/1000:.1f}B'))
ax.set_xlim(1, 35)
ax.set_xticks([1, 2, 3, 5, 10, 20, 30])
ax.set_ylim(85.5, 90.9)
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax.grid(True, which='both', linestyle='--', linewidth=0.3, zorder=0)

sizes_legend = [0.8, 2.0, 3.4]
labels_legend = ['0.8B', '2.0B', '3.4B']

legend_ax = fig.add_axes([0.55, 0.2, 0.42, 0.3])  # [left, bottom, width, height]
legend_ax.axis('off')

for i, (size, label) in enumerate(zip(sizes_legend, labels_legend)):
    x = 0.17 + 0.33 * i
    legend_ax.scatter(x, 0.56, s=flops_scale(size), c='gray', alpha=0.6, edgecolors='black')
    legend_ax.text(x, 0.1, label, va='center', ha='center', fontsize=8)

legend_ax.text(0.5, 1.1, 'FLOPs(B)', fontweight='bold', fontsize=9, va='center', ha='center')

rect = patches.Rectangle((0, 0), 1, 1, fill=False, edgecolor='gray', linewidth=1)
legend_ax.add_patch(rect)

plt.tight_layout()
plt.savefig('3DSGG_triacc.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker
import matplotlib.patches as patches

#    
new_data = {
    'Model': ['SGPN\n(2020)', 'SGFN\n(2021)', 'SGG point\n(2021)', 'SGFN+Attn\n(2023)', 'VL-SAT\n(2023)'],
    'Total_Params': [3.29, 2.90, 1.92, 8.20, 25.06],  # : 
    'GNN_Params': [2.63, 2.25, 0.59, 7.38, 21.35],    # : 
}
df_new = pd.DataFrame(new_data)
df_new['Year'] = df_new['Model'].str.extract('(\d{4})').astype(int)
df_new = df_new.sort_values(['Year', 'Model'], ascending=[True, True])

plt.style.use('seaborn-whitegrid')
plt.rcParams['font.family'] = 'serif'

fig, ax1 = plt.subplots(figsize=(9, 4))

bar_height = 0.35
index = range(len(df_new))
total_bars = ax1.barh(index, df_new['Total_Params'], bar_height, label='Total Parameters', color='#4C72B0', alpha=0.8)
gnn_bars = ax1.barh([i + bar_height for i in index], df_new['GNN_Params'], bar_height, label='GNN Parameters', color='#55A868', alpha=0.8)

for i, (total, gnn) in enumerate(zip(df_new['Total_Params'], df_new['GNN_Params'])):
    ax1.text(total, i, f' {total:.2f}M ', va='center', ha='left', fontweight='bold', fontsize=10, color='#4C72B0')
    ax1.text(gnn, i + bar_height, f' {gnn:.2f}M ', va='center', ha='left', fontweight='bold', fontsize=10, color='#55A868')

ax1.set_yticks([i + bar_height/2 for i in index])
ax1.set_yticklabels(df_new['Model'], fontsize=10)
ax1.set_xlabel('Number of Parameters (Millions)', fontsize=10)
ax1.set_ylabel('Model', fontsize=12)

ax1.set_xlim(0, max(df_new['Total_Params']) * 1.1)  
ax1.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax1.xaxis.set_minor_locator(ticker.MultipleLocator(1))

ax1.grid(True, which='major', linestyle='--', linewidth=0.5, axis='x')
ax1.grid(True, which='minor', linestyle=':', linewidth=0.3, axis='x')

ax1.invert_yaxis()

legend_box = patches.Rectangle((0.66, 0.855), 0.3, 0.135, fill=False, transform=ax1.transAxes, clip_on=True, linewidth=0.37)
ax1.add_patch(legend_box)

ax1.text(0.68, 0.955, '', color='#4C72B0', transform=ax1.transAxes, fontsize=18, va='center')
ax1.text(0.72, 0.95, 'Total Parameters', transform=ax1.transAxes, fontsize=15, va='center')
ax1.text(0.68, 0.885, '', color='#55A868', transform=ax1.transAxes, fontsize=18, va='center')
ax1.text(0.72, 0.88, 'GNN Parameters', transform=ax1.transAxes, fontsize=15, va='center')

#    
plt.tight_layout()
# plt.savefig('3DSGG_models_params_final.png', bbox_inches='tight', dpi=300)
plt.show()
```
